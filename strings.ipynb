#### SLICING

```python
#Used to access many characters in a string unlike indexing where you can only access one at a time
```

```python
city = "NAIROBI"
```

```python
print(city[0:3])
#grab characters from zero upto but not including 3
```

```python
print(city[3:])
#grab from char 3 to the last
```
