### "If" STATEMENT

```python
# gives only choice,it executes when True and ignores the code when false
```

```python
if 5>3:
    print("A")
```

```python
age = int(input("Please enter your age:"))
if age >= 18:
    print("You are allowed to vote!!!")
```

```python
gender = input("Gender:")
if gender== "Male":
    print("Get inside!")
```

### "if" "else" statement

```python
# gives two choices,if True and if False
```

```python
height = float(input("height"))
if height <= 149.0:
    print("You are a dwarf")
else:
    print("You are not a dwarf!")
```

```python
age = int(input("Enter your age:"))
if age >= 35:
    print("You can vye for Presidency!!")
else:
    print("Sorry You can't vye for presidency!!")
```

### "if...elif...else"

```python
#used when you have multiple conditions
#Python reads them from top to bottom.
```

```python
score = float(input("Enter your score:"))
if score >= 80:
    print("Grade:A")
elif score >= 60:
    print("Grade:B")
elif score >= 50:
    print("Grade:C")
elif score >= 40:
    print("Grade:D")
else:
    print("SUPPLEMENTARY!!")
```

```python
language = input("Enter National Language")
if language == "Swahili":
    print("You are from Tanzania")
elif language == "Kinywarwanda":
    print("You are from Rwanda")
elif language == "French":
    print("You are from Senegal")
elif language == "English":
    print("You are from the UK")
else:
    print("Tell me your country")
```

### NESTED CONDITIONAL STATEMENTS

```python
# a condition inside a statement
```

```python
# we use it to make more sensible conditions
```

```python
#In the above example,a value less than 0 ie -10 is Printed out as supp,we therefore use nested if to make it logically sensible
```

```python
score = float(input("Enter your score:"))
if 0 <= score <= 100:
    if score >= 80:
        print("Grade:A")
    elif score >= 60:
        print("Grade:B")
    elif score >= 50:
        print("Grade:C")
    elif score >= 40:
        print("Grade:D")
    else:
        print("SUPPLEMENTARY!!")
else:
    print("Enter a valid Score")
```

```python
# example 2 of nested conditional statements
```

```python
correct_pin = 1234
entered_pin = float(input("Please enter MPESA PIN"))
balance = 30000
withdraw_amount = float(input("Please enter AMOUNT you wish to withdraw"))
#first if statement
if entered_pin == correct_pin:
    print("PIN accepted!")
    if withdraw_amount <= balance:
        print(withdraw_amount, "Has been succesfully sent to your MPESA")
    else:
        print("Insufficient Amount!!")    
else:
    print("Incorrect PIN try again")
```
