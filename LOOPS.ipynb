### 1. "While" loop

```python
#while loop keeps running if condition is True and stops when False
```

```python
# Number of passengers in a car,as long as number is below 13 continue
number_of_passengers = 0
while number_of_passengers <= 14:
    print(f"Passengers {number_of_passengers} boarded")
    number_of_passengers = number_of_passengers + 1
    print("Gari imejaa Todhie")
```

```python
# lets say continue adding students until max of 40 in a class
students = 0
while students <= 40:
    print(f"student {students} has entered the class")
    students = students + 1
```

```python
# saving 1k until 100k
savings = 0
while savings < 100000:
    savings = savings + 1000
    print(f"savings is now {savings}")
```

### 2. "for" loop

```python
# for loop is used to itearate over a sequence ie list,tuple,string,range
```

```python
# for loop in a list
counties = ["Nairobi","Kisumu","Eldoret","Mombasa"]
for county in counties:
    print(f"I want to visit {county}")
```

```python
# Letters in a string
word = "TANZANIA"
for letter in word:
    print(letter)
```

### 3. Range()

```python
# it allows us to loop a specific number of times
```

```python
# let's print from 0 to 9
print("---------Standard Range------")
for i in range(10):
    print(i)
```

```python
#lets print from 0 to 100
print("-----Standard range")
for i in range(100):
    print(i)
```

```python
# start at 3 stop before 12
print("--------Start at 3 stop just before 12-------")
for i in range(3,12):
    print(i)
```

```python
# from 0 to 27 step of 3
print("------Standard Range------")
for i in range(0,30,3):
    print(i)
```

```python
# from 0 to 80 steps of 2
print("------Standard range--------")
for i in range(0,82,2):
    print(i)
```

#### 4.1 Break statement

```python
# the break statement terminates the loop entirely
```

```python
for num in range(1,10):
    print(num)
    if num == 6:
        print("We are at 5! stop loop")
        break 
```
