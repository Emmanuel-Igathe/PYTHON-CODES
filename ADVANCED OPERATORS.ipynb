### 1.Comparison Operators

```python
# comparison operators compare two values, i.e., is it equal to? or is it bigger than?
```

```python
# examples of comparison operators
# 1.  == Equal to
# 2. =! Not equal to
# 3. >= Greater than or equal to
# 4. <= Less than or equal to
# 5. > Greater than
# 6. < Less than
```

```python
# Buying lunch and checking M-Pesa balance
balance  = int(input("Enter M-PESA balance: "))
cost_of_lunch = 650
if balance >= cost_of_lunch :
    print("Yes you can afford Lunch!")
else:
    print("Oyaah utachonga Waru")
```

```python
# Using the teachers example
age = int(input("Please enter your age: "))
is_youth = 18 <= age <= 35
print(f"is youth", {is_youth})
```

### 2.LOGICAL OPERATORS

```python
# We use logical operators when we need to check more than one condition at a time ,i.e , AND,OR ,NOT
```

##### AND OPERATOR

```python
has_id = True
has_fee = True
can_enter = has_id and has_fee
print(can_enter)
```

```python
# if a person can vote
has_id = input("True/False")
has_registered_as_voter = input("True/False")
can_vote = has_id and has_registered_as_voter
print(can_vote)
```

##### OR OPERATOR

```python
# you can continue if at least one of the conditions is met
```

```python
# to enter school
official_clothing = input("True/False")
well_groomed = input("True/False")
can_enter_school = official_clothing or well_groomed 
print(can_enter_school)
```

```python
# To enter the library
is_student = input("True/False")
is_staff = input("True/False")
can_enter_library = is_student or is_staff
print(can_enter_library)
```

##### NOT OPERATOR

```python
# it is used to reverse the result ie True becomes False
```

```python
is_underage = False
is_allowed = not is_underage
print(is_allowed)
```

# IDENTITY OPERATORS

```python
# Identity operators check if two variables refer to the same object in the memory,not just if they have the same value
```

```python
# They include 'is' and 'is not'
```

```python
# two separate lists with same content
list_A = [1, 2, 3]
list_B = [1, 2, 3]
list_C = list_A
print(list_C)
```

```python
# comparing values
print(list_A == list_B)
```

```python
# comparing identity
print(list_A is list_B)  #false because they are two different lists in the memory
```

```python
print(list_A is list_C)
```

## MEMBERSHIP OPERATORS

```python
# membership operators check if a specific item exists
menu = ['chapati','pilau','githeri','matoke']
print('chapati' in menu)  # is True because chapati is in the menu
```

```python
print('biriani' in menu) # is False because biriani is not in the menu
```
